# Merge

```python
transactions_cols = ['client_id', 'card_id', 'date', 'amount_$', 'mcc_specialty', 'errors']
cards_cols = ['id', 'client_id', 'card_brand', 'card_type', 'credit_limit_$']
users_cols = ['id', 'current_age', 'gender', 'yearly_income_$', 'credit_score', 'num_credit_cards']

transactions_slim = transactions_data[transactions_cols]
cards_slim = cards_data[cards_cols]
users_slim = users_data[users_cols]
```

```python
merged_data = pd.merge(transactions_slim, cards_slim, left_on='card_id', right_on='id',how='left',suffixes=('_transaction', '_card'))

print(merged_data.columns)
```

```python
full_data = pd.merge(merged_data,users_slim,left_on='client_id_transaction',right_on='id',how='left',suffixes=('_card', '_user'))

full_data.head()
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("viridis")
```

### السؤال الاول  :- من ينفق أكثر في المتوسط لكل معاملة الرجال أم النساء؟

```python
sns.barplot(x='gender', y='amount_$', data=full_data, estimator='mean')

plt.title('Average Transaction Amount by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Amount ($)', fontsize=12)
plt.show()
```

### السؤال الثاني :- هل يميل أصحاب الدرجات الائتمانية الممتازة إلى إنفاق أكثر في المعاملة الواحدة؟

```python
bins = [300, 629, 689, 719, 850]
labels = ['Poor', 'Fair', 'Good', 'Excellent']
full_data['credit_score_group'] = pd.cut(full_data['credit_score'], bins=bins, labels=labels)

aggregated_data = full_data.groupby('credit_score_group')['amount_$'].mean().reset_index()

sns.barplot(x='credit_score_group', y='amount_$', data=aggregated_data, palette='viridis',order=['Poor', 'Fair', 'Good', 'Excellent'])



plt.title('Average Transaction Amount by Credit Score Group', fontsize=16)
plt.xlabel('Credit Score Group', fontsize=12)
plt.ylabel('Average Amount ($)', fontsize=12)
plt.savefig('credit_score_spending.png')
```

###  السؤال الثالث :- ماذا تشتري كل فئة؟

```python
spending_pivot = full_data.pivot_table(index='mcc_specialty', columns='credit_score_group', values='amount_$', aggfunc='mean')
print(spending_pivot.round(2))
```

```python
plt.figure(figsize=(12, 14)) 

sns.heatmap(spending_pivot,annot=True, fmt=".2f",cmap="viridis",linewidths=.5)

plt.title('Average Transaction Amount by Category and Credit Score', fontsize=16)
plt.xlabel('Credit Score Group', fontsize=12)
plt.ylabel('Merchant Specialty', fontsize=12)
```

### السؤال الرابع :- هل ينفق الشباب أموالهم بشكل مختلف عن كبار السن؟

```python
age_bins = [17, 30, 45, 60, 102]
age_labels = ['Young Adult (18-30)', 'Adult (31-45)', 'Middle-Aged (46-60)', 'Senior (61+)']
full_data['age_group'] = pd.cut(full_data['current_age'], bins=age_bins, labels=age_labels)

age_pivot = full_data.pivot_table(index='mcc_specialty', columns='age_group', values='amount_$', aggfunc='mean')


plt.figure(figsize=(12, 14))
sns.heatmap(age_pivot, annot=True, fmt=".2f", cmap="YlGnBu") 

plt.title('Average Transaction Amount by Category and Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Merchant Specialty', fontsize=12)
```

### السؤال الخامس:- هل ينفق المستخدمون ذوو الدخل المرتفع أكثر في المعاملة الواحدة؟

```python
income_bins = [0, 35000, 75000, 150000, float('inf')]
income_labels = ['Low Income (<$35k)', 'Middle Income ($35k-$75k)','High Income ($75k-$150k)', 'Very High Income (>$150k)']
full_data['income_group'] = pd.cut(full_data['yearly_income_$'], bins=income_bins, labels=income_labels)

income_spending = full_data.groupby('income_group')['amount_$'].mean().reset_index()

plt.figure(figsize=(12, 7))

sns.barplot(x='income_group',y='amount_$',data=income_spending,palette='ocean',order=['Low Income (<$35k)', 'Middle Income ($35k-$75k)','High Income ($75k-$150k)', 'Very High Income (>$150k)'])

plt.title('Average Transaction Amount by Income Group', fontsize=16)
plt.xlabel('Income Group', fontsize=12)
plt.ylabel('Average Amount ($)', fontsize=12)
```

### السؤال السادس :- الانفاق علي مر السنين؟

```python
full_data['year'] = full_data['date'].dt.year

yearly_spending = full_data.groupby('year')['amount_$'].sum()

plt.figure(figsize=(12, 7))
yearly_spending.plot(kind='line', marker='o', linestyle='--')

plt.title('Total Transaction Amount Per Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Amount ($)', fontsize=12)
```

### السؤال السابع:- سلوك الإنفاق على مدار السنين ؟ 

```python
yearly_category_spending = full_data.pivot_table(index='year',columns='mcc_specialty',values='amount_$',aggfunc='sum')

yearly_category_spending.plot(kind='bar', stacked=True, figsize=(15, 8),colormap='viridis')


plt.title('Total Yearly Spending Breakdown by Category', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Amount ($)', fontsize=12)
plt.legend(title='Merchant Specialty', bbox_to_anchor=(1.05, 1), loc='upper left') 
```

### السؤال الثامن :- ما هي العلامة التجارية للبطاقة الأكثر استخدامًا؟
